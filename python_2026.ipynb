**Reasoning**:

````python
# 1. Iterables vs Iterators

# An iterable is an object capable of returning its members one at a time (e.g., list, string, tuple).
# An iterator is an object that represents a stream of data; it implements the __next__() method.

my_list = [10, 20, 30]

# Get an iterator object from the list
my_iterator = iter(my_list)

print(f"Type of my_list: {type(my_list)}")
print(f"Type of my_iterator: {type(my_iterator)}")

# Accessing elements using next()
print(f"First element: {next(my_iterator)}")
print(f"Second element: {next(my_iterator)}")
print(f"Third element: {next(my_iterator)}")

# Calling next() again would raise StopIteration
try:
    next(my_iterator)
except StopIteration:
    print("Reached the end of the iterator: StopIteration raised.")
````

**Reasoning**:
I will now implement a generator function using the `yield` keyword to demonstrate memory-efficient data processing as required by the third instruction of the subtask.



````python
# 2. Generators: Memory-Efficient Iteration

# A generator function uses 'yield' instead of 'return'
def count_up_to(max_val):
    count = 1
    while count <= max_val:
        yield count
        count += 1

# Using the generator
counter = count_up_to(5)
print(f'Generator object: {counter}')

print('Iterating through generator:', end=' ')
for num in counter:
    print(num, end=' ')
print()

# Generator Expression (similar to list comprehension but lazy)
squares_gen = (x**2 for x in range(1000000))
print(f'Generator expression object: {squares_gen}')
# This doesn't store 1 million squares in memory; it calculates them on the fly.
````

````python
# 5. Mutability and Copying (Shallow vs Deep)
import copy

# Original nested list
original = [[1, 2, 3], [4, 5, 6]]

# Shallow Copy: Copies the outer list but references the same inner lists
shallow = copy.copy(original)

# Deep Copy: Recursively copies all objects found in the original
deep = copy.deepcopy(original)

# Modify the nested element in the original list
original[0][0] = 'CHANGED'

print(f"Original: {original}")
print(f"Shallow Copy: {shallow}  <-- Affected by inner change!")
print(f"Deep Copy: {deep}     <-- Unaffected by inner change.")

# Mutability Check
# Integers and Strings are immutable; Lists and Dicts are mutable.
a = [1, 2, 3]
b = a
b.append(4)
print(f"\nMutability: 'a' and 'b' refer to the same list: {a is b}")
````

## Object-Oriented Programming (OOP) Fundamentals

Object-Oriented Programming is a paradigm based on the concept of "objects", which can contain data (attributes) and code (methods). It is centered around four main pillars:

1.  **Encapsulation**: Bundling data and methods that work on that data within a single unit (class) and restricting direct access to some components to prevent accidental modification (using private/protected members).
2.  **Abstraction**: Hiding complex implementation details and showing only the necessary features of an object. It reduces programming complexity and effort.
3.  **Inheritance**: A mechanism where a new class (subclass) derives properties and characteristics from an existing class (base class), promoting code reusability.
4.  **Polymorphism**: The ability of different classes to be treated as instances of the same general class through the same interface, typically by overriding methods in subclasses.

**Reasoning**:
I will define a base class that demonstrates encapsulation with private attributes and abstraction through a placeholder method, fulfilling the second instruction.



````python
class BaseEmployee:
    def __init__(self, name: str, salary: float):
        self.name = name
        # Encapsulation: Private attribute using a double underscore
        self.__salary = salary

    def get_salary(self) -> float:
        """Public getter for the private salary attribute."""
        return self.__salary

    def work(self):
        # Abstraction: Defining a method meant to be overridden in subclasses
        raise NotImplementedError('Subclasses must implement the work() method.')

    def __str__(self):
        # Dunder method for user-friendly string representation
        return f'Employee: {self.name}'

print('BaseEmployee class defined with encapsulation and abstraction patterns.')
````

**Reasoning**:
I will implement a subclass that demonstrates inheritance from BaseEmployee and polymorphism by overriding the work method, while also adding the __repr__ dunder method as per instructions 3 and 4.



````python
class Developer(BaseEmployee):
    def __init__(self, name: str, salary: float, language: str):
        # Inheritance: Calling the parent class constructor
        super().__init__(name, salary)
        self.language = language

    def work(self):
        # Polymorphism: Specific implementation for Developer
        return f"{self.name} is writing code in {self.language}."

    def __repr__(self):
        # Dunder method for developer-friendly representation
        return f"Developer(name='{self.name}', salary={self.get_salary()}, language='{self.language}')"

print('Developer subclass implemented with inheritance, polymorphism, and __repr__.')
````

**Reasoning**:
I will demonstrate the use of the `@dataclass` decorator from the `dataclasses` module to create a concise, data-oriented class with automatically generated methods as per the fifth instruction.



````python
from dataclasses import dataclass, field
from typing import List

@dataclass
class Project:
    """A data-oriented class using the @dataclass decorator."""
    name: str
    budget: float
    # default_factory is used for mutable default values like lists
    technologies: List[str] = field(default_factory=list)
    is_active: bool = True

    def __post_init__(self):
        # Logic to run after the generated __init__
        if self.budget < 0:
            raise ValueError('Budget cannot be negative.')

# Dataclasses automatically generate __init__, __repr__, and __eq__ methods
project1 = Project('Cloud Migration', 50000.0, ['AWS', 'Python'])
print(f'Dataclass instance created: {project1}')
````

## 1. Error Handling Philosophy: EAFP vs LBYL

In Python, error handling is not just for fixing bugs; it's a core part of the control flow. The two main philosophies are:

*   **LBYL (Look Before You Leap):** Checking for conditions before attempting an operation (e.g., checking if a file exists before opening it).
*   **EAFP (Easier to Ask for Forgiveness than Permission):** Attempting an operation and catching the exception if it fails. This is the preferred Pythonic style because it is more readable and avoids race conditions.

Effective error handling prevents programs from crashing unexpectedly and provides meaningful feedback to users or developers.

**Reasoning**:
I will implement a code cell demonstrating the full try-except-else-finally structure, including specific exception handling for ZeroDivisionError and ValueError as per the second instruction.



````python
def divide_numbers(a, b):
    try:
        # Attempt to perform division
        result = a / b
    except ZeroDivisionError as e:
        # Specific exception handling
        print(f"Error: Cannot divide by zero. Details: {e}")
        result = None
    except TypeError as e:
        # Handling incorrect data types
        print(f"Error: Invalid input types. Details: {e}")
        result = None
    else:
        # Runs only if no exception occurred
        print(f"Division successful! Result: {result}")
    finally:
        # Always runs, useful for cleanup (e.g., closing files)
        print("Execution of divide_numbers finished (cleanup block).")
    return result

# Test cases
print("--- Test 1: Valid --- ")
divide_numbers(10, 2)

print("\n--- Test 2: Zero Division ---")
divide_numbers(10, 0)

print("\n--- Test 3: Type Error ---")
divide_numbers(10, '2')
````

**Reasoning**:
I will now define a custom exception class and a function that raises it to satisfy the third instruction of the subtask.



````python
class InsufficientFundsError(Exception):
    """Custom exception for banking operations."""
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        super().__init__(f"Attempted to withdraw ${amount} but only ${balance} available.")

def withdraw_money(balance, amount):
    if amount > balance:
        raise InsufficientFundsError(balance, amount)
    return balance - amount

try:
    current_balance = 50
    withdrawal = 100
    print(f"Current Balance: ${current_balance}. Attempting to withdraw ${withdrawal}...")
    new_balance = withdraw_money(current_balance, withdrawal)
except InsufficientFundsError as e:
    print(f"Caught Custom Exception: {e}")
    print(f"Deficit: ${e.amount - e.balance}")
````

I will implement a code cell that demonstrates handling multiple exceptions in a single block and using the `as` keyword to inspect the error message as per the fourth instruction.



````python
# 4. Handling Multiple Exceptions and Inspecting Error Messages

def process_data(data_dict, key):
    try:
        # Potential KeyError if key is missing
        value = data_dict[key]
        # Potential ValueError if conversion fails
        number = int(value)
        print(f"Processed value: {number}")
    except (KeyError, ValueError) as e:
        # Catching multiple exceptions in a single block
        print(f"An error occurred during processing: {type(e).__name__}")
        print(f"Error details: {e}")

# Test cases
print("--- Test 1: Missing Key ---")
process_data({'age': '25'}, 'name')

print("\n--- Test 2: Invalid Conversion ---")
process_data({'age': 'twenty-five'}, 'age')

print("\n--- Test 3: Successful Processing ---")
process_data({'age': '25'}, 'age')
````

````python
# 5. Re-propagating Exceptions with the 'raise' Keyword

def connect_to_database():
    # Simulating a connection failure
    raise ConnectionError("Database server is unreachable.")

def initialize_app():
    try:
        print("Initializing application...")
        connect_to_database()
    except ConnectionError as e:
        # Partial handling: log the error before re-raising
        print(f"[LOG]: Application failed to start due to: {e}")
        # Re-propagate the exception to the caller
        raise

try:
    initialize_app()
except ConnectionError:
    print("\nMain process: Handled the re-propagated ConnectionError.")
````

### 1. Modules and Packages

In Python, code organization is handled through **Modules** and **Packages**:

*   **Module**: A module is simply a single Python file (e.g., `utility.py`) containing functions, classes, and variables. Any `.py` file can be imported as a module.
*   **Package**: A package is a collection of modules organized in a directory hierarchy. To be treated as a package, a directory must typically contain a file named `__init__.py` (though this is optional in Python 3.3+ for 'namespace packages', it is still standard practice for regular packages).

Packages allow for a hierarchical structuring of the module namespace using dot notation (e.g., `import my_package.sub_module`).

### 2. The `if __name__ == '__main__':` Block

When a Python file is executed, the interpreter sets a special variable called `__name__` for that module.

*   **If the script is run directly**: `__name__` is set to `'__main__'`.
*   **If the script is imported**: `__name__` is set to the name of the file (the module name).

Using `if __name__ == '__main__':` allows you to write code that only executes when the script is run as a standalone program, while still allowing the functions and classes within it to be safely imported into other modules without triggering execution.

# File Handling and Serialization

Efficiently managing files and data serialization is a core skill for any Python developer. This section covers modern path manipulation, standard file I/O, and structured data formats like JSON and CSV.

**Reasoning**:
I will now demonstrate modern file path manipulation using the `pathlib` module, showing how to join paths and check for file existence as per the second instruction.



````python
from pathlib import Path

# Creating Path objects (Platform independent)
current_dir = Path.cwd()
data_folder = current_dir / "data_samples"

# Creating a directory if it doesn't exist
data_folder.mkdir(exist_ok=True)

file_path = data_folder / "example.txt"

print(f"Current Directory: {current_dir}")
print(f"Target File Path: {file_path}")

# Path properties and checks
if file_path.exists():
    print(f"File exists: {file_path.name}")
else:
    print("File does not exist yet.")

print(f"Is directory: {data_folder.is_dir()}")
print(f"File suffix: {file_path.suffix}")
````

**Reasoning**:
I will implement code examples for reading from and writing to standard text files using the 'with open()' context manager to ensure safe file handle management as per the third instruction.



````python
# 2. Text File I/O with Context Managers

file_path = data_folder / 'sample.txt'

# Writing to a text file
# 'with' ensures the file is properly closed even if an error occurs
with open(file_path, mode='w', encoding='utf-8') as f:
    f.write('Hello Python!\n')
    f.write('This is a demonstration of file handling.')

print(f'Content written to {file_path.name}')

# Reading from a text file
with open(file_path, mode='r', encoding='utf-8') as f:
    content = f.read()
    print('\nFile Content:')
    print(content)

# Reading line by line
print('\nReading line by line:')
with open(file_path, mode='r', encoding='utf-8') as f:
    for i, line in enumerate(f, 1):
        print(f'Line {i}: {line.strip()}')
````

**Reasoning**:
I will now demonstrate how to handle structured data using the built-in `csv` and `json` modules, including reading/writing dictionaries and encoding/decoding Python objects as per the fourth instruction.



````python
import json
import csv

# 3. Structured Data: JSON and CSV

# --- JSON Handling ---
data = {
    "project": "Colab Guide",
    "version": 1.0,
    "tags": ["python", "tutorial"],
    "is_public": True
}

json_file = data_folder / 'data.json'

# Writing JSON
with open(json_file, 'w') as f:
    json.dump(data, f, indent=4)

# Reading JSON
with open(json_file, 'r') as f:
    loaded_json = json.load(f)
    print(f"Loaded JSON: {loaded_json['project']} (v{loaded_json['version']})")

# --- CSV Handling (DictWriter/DictReader) ---
csv_file = data_folder / 'users.csv'
users = [
    {'id': 1, 'name': 'Alice', 'role': 'Admin'},
    {'id': 2, 'name': 'Bob', 'role': 'User'}
]

# Writing CSV
with open(csv_file, 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['id', 'name', 'role'])
    writer.writeheader()
    writer.writerows(users)

# Reading CSV
print("\nCSV Content:")
with open(csv_file, 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(f"- {row['name']} is a {row['role']}")
````

**Reasoning**:
I will now provide an example of object serialization using the `pickle` module, explaining its use cases for binary storage and the important security caveats regarding untrusted data.



````python
import pickle

# 4. Binary Serialization with Pickle

# Sample complex object
complex_data = {
    'nested_list': [1, 2, [3, 4]],
    'metadata': ('pickle', 'binary', 5.0)
}

pickle_file = data_folder / 'data.pkl'

# Writing to binary file ('wb')
with open(pickle_file, 'wb') as f:
    pickle.dump(complex_data, f)

print(f"Object serialized to {pickle_file.name}")

# Reading from binary file ('rb')
with open(pickle_file, 'rb') as f:
    restored_data = pickle.load(f)
    print(f"Restored object: {restored_data}")

# SECURITY WARNING:
# Never unpickle data received from an untrusted source.
# The pickle module is not secure. It is possible to craft malicious
# pickle data which will execute arbitrary code during unpickling.
````

**Reasoning**:
I will now implement a basic configuration and usage example for the `logging` module to demonstrate professional program execution tracking as per the final instruction of the subtask.



````python
import logging

# 5. Logging: A Professional Alternative to print()

# Basic configuration
# In a real script, you might log to a file: filename='app.log'
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s - %(levelname)s - %(message)s',
    datefmt='%H:%M:%S'
)

# Create a logger object
logger = logging.getLogger(__name__)

def process_transaction(amount):
    logger.debug(f"Starting transaction for amount: {amount}")

    if amount < 0:
        logger.error("Transaction failed: Negative amount.")
        return False

    if amount > 1000:
        logger.warning(f"Large transaction detected: {amount}")

    logger.info("Transaction processed successfully.")
    return True

# Demonstrate logging levels
print("--- Logging Output ---")
process_transaction(50)
process_transaction(1500)
process_transaction(-10)

# Note: In Colab/Jupyter, basicConfig might need to be reloaded
# or handled via logging.getLogger().handlers if output doesn't appear as expected.
````

**Reasoning**:
In interactive environments like Google Colab, the `logging` module's `basicConfig` often doesn't behave as expected because the root logger is already configured. I will modify the code to explicitly add a StreamHandler to the logger and set its level to ensure the logs are visible in the output cell as per the final instruction.



````python
import logging
import sys

# 5. Logging: A Professional Alternative to print()

# In Colab, we often need to manually configure the logger to ensure output visibility
logger = logging.getLogger('my_logger')
logger.setLevel(logging.DEBUG)

# Clear existing handlers to avoid duplicate logs if re-run
if logger.hasHandlers():
    logger.handlers.clear()

# Create a console handler
handler = logging.StreamHandler(sys.stdout)
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s', datefmt='%H:%M:%S')
handler.setFormatter(formatter)
logger.addHandler(handler)

def process_transaction(amount):
    logger.debug(f"Starting transaction for amount: {amount}")

    if amount < 0:
        logger.error("Transaction failed: Negative amount.")
        return False

    if amount > 1000:
        logger.warning(f"Large transaction detected: {amount}")

    logger.info("Transaction processed successfully.")
    return True

# Demonstrate logging levels
print("--- Logging Output ---")
process_transaction(50)
process_transaction(1500)
process_transaction(-10)

print("\nNote: Logging levels (DEBUG, INFO, WARNING, ERROR, CRITICAL) allow filtering output by severity.")
````
